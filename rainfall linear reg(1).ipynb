```python
data5=data4.groupby(["YEAR"]).count()
```

```python
display(data5)
```

```python
data6=a.get_dummies(data4,dtype=int)
```

```python
display(data6)
```

```python
y=data6['ANNUAL']
x=data6.drop('ANNUAL',axis=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
ypred=reg.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error 
mean_squared_error(ypred,y_test)
```

```python
Results=a.DataFrame(columns=['ANNUAL','PREDICTED'])
Results['ANNUAL']=y_test
Results['PREDICTED']=ypred
Results=Results.reset_index()
Results['YEAR']=Results.index
Results.head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='YEAR',y='ANNUAL',data=Results.head(15))
sns.lineplot(x='YEAR',y='PREDICTED',data=Results.head(15))
plt.plot()
```
